```python
import numpy as np
import torch

import matplotlib.pyplot as plt

# sympy = symbolic math in Python
import sympy as sym
import sympy.plotting.plot as symplot
```

## Dot Product

```python
# create a vector
nv1 = np.array([1,2,3,4])
nv2 = np.array([0,1,0,-1])

# dot product via function
print(np.dot(nv1,nv2))

# dot product via computation
print(np.sum( nv1*nv2 ))


print("--"*10)


# create a vector
tv1 = torch.tensor([1,2,3,4])
tv2 = torch.tensor([0,1,0,-1])

# dot product via function
print(torch.dot(tv1,tv2))

# dot product via computation
print(torch.sum( tv1*tv2 ))
     
```

## Max/Min & ArgMin/ArgMax

```python
v = np.array([1,40,2,-3])

minval = np.min(v)
maxval = np.max(v)

mininidex = np.argmin(v)
maxindex = np.argmax(v)



print(f"Minval: {minval} \nMaxval: {maxval} \nMinIndex: {mininidex} \nMaxIndex: {maxindex}")


print("--"*10)

v = torch.tensor([1,40,2,-3])
minval = torch.min(v)
maxval = torch.max(v)

mininidex = torch.argmin(v)
maxindex = torch.argmax(v)


print(f"Minval: {minval} \nMaxval: {maxval} \nMinIndex: {mininidex} \nMaxIndex: {maxindex}")
```

## Derivatives

```python
# create symbolic variables in sympy
x = sym.symbols('x')

def plot(fx,df):
    p1 = symplot(fx, (x, -4, 4), line_color='blue', show=False, legend=True)
    p2 = symplot(df, (x, -4, 4), line_color='red', show=False, legend=True)

    # Add legends
    p1[0].label = 'f(x)'
    p2[0].label = "f'(x)"

    # Combine and show
    p1.extend(p2)
    p1.title = 'Function and its Derivative'
    p1.show()
```

```python
# create a function
fx = 2*x**2

# compute its derivative
df = sym.diff(fx,x)

# print them
print(fx)
print(df)

plot(fx,df)
```

```python
# repeat with relu and sigmoid

# create symbolic functions
relu = sym.Max(0,x)
sigmoid = 1 / (1+sym.exp(-x))

# graph the functions
print("Relu")
plot(relu,sym.diff(relu))


# graph their derivatives
print("Sigmoid")
plot(sigmoid,sym.diff(sigmoid))
```

## Entropy

```python
# probability of an event happening
p = .25

# NOT the correct formula!
H = -( p*np.log(p) )
print('Wrong entropy: ' + str(H))
```

```python
## Cross Entropy

# note: all probs must sum to 1!
p = [   1,0   ] # sum=1
q = [ .25,.75 ] # sum=1

H = 0
for i in range(len(p)):
  H -= p[i]*np.log(q[i])

print('Cross entropy: ' + str(H))
```
